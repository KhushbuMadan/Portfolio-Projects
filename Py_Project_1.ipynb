```python
import matplotlib.pyplot as plt

#Data to Plot
Labels=['EUBS','TUM','LMU','MBS']
Sizes = [33,27,22,18]

#Create a Pie
plt.pie (Sizes,labels=Labels)

#Title
plt.title ('Business Schools in Munich')

#Legend
plt.legend(loc='best')

# Save the plot as an Image
plt.savefig('Business_Schools_Pie.png')

#Show Pie-chart
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data to Plot
Labels = ['EUBS', 'TUM', 'LMU', 'MBS']
Sizes = [33, 27, 22, 18]
explode = [0.1, 0.1, 0.1, 0.1]

# Define colors
colors = ['#ff0000','#0000ff','#00ff00','#ff9999']

# Create a Pie
plt.pie(Sizes, labels=Labels, autopct='%1.1f%%', colors=colors, shadow=True)

# Title
plt.title('Business Schools in Munich')

# Legend
plt.legend(loc=(0.9,0.8))

# Save the plot as an Image
plt.savefig('Business_Schools_Pie.png')

# Show Pie-chart
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data to Plot
Labels = ['EUBS', 'TUM', 'LMU', 'MBS']
Sizes = [33, 27, 22, 18]

# Create a Doughnut Chart
plt.pie(Sizes, labels=Labels, autopct='%1.1f%%', wedgeprops={'width': 0.3})

# Title
plt.title('Business Schools in Munich')

# Legend
plt.legend(loc=(0.9,0.8))

# Save the plot as an Image
plt.savefig('Business_Schools_Doughnut.png')

# Show Doughnut Chart
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data to Plot
Labels = ['EUBS', 'TUM', 'LMU', 'MBS']
Sizes = [30000, 27000, 22000, 18000]
colors = ['#ff0000', '#0000ff', '#00ff00', '#ff9999']

# Create a Bar Chart
plt.bar(Labels, Sizes, color=colors)

# Title
plt.title('Business Schools in Munich')

# Add labels to the axes
plt.xlabel('Schools')
plt.ylabel('Number of Students')

# adding labels to each bar
for i, v in enumerate(Sizes):
   plt.text(i, v + 500, str(v), ha='center')

# Show Bar Chart
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data to Plot
labels= ['EUBS','TUM','LMU','MBS']
sizes= [30000,27000,22000,18000]

# Create a Step Chart
plt.step(labels,sizes)

# Add Title & labels to the axes
plt.title('Business schools Munich')
plt.xlabel('No. of Students')
plt.ylabel('Business Schools')

# Show Step Chart
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data to Plot
labels= ['EUBS','TUM','LMU','MBS']
sizes= [30000,27000,22000,18000]

# Create a Stem Chart
plt.stem(labels,sizes)

# Add Title & labels to the axes
plt.title('Business schools Munich')
plt.xlabel('No. of Students')
plt.ylabel('Business Schools')

# adding labels to each Branch
for i, v in enumerate(Sizes):
   plt.text(i, v + 500, str(v), ha='center')

# Show Stem Chart
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# To Generate random data
randn = np.random.randn(1000)

# To Create a histogram
plt.hist(randn, bins=30, color='green', edgecolor='white')

# Add title and labels
plt.title('Histogram of Random Data')
plt.xlabel('Values')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data to plot
category_1= [20,40,60,80,100]
category_2= [10,30,50,70,90]
category_3= [5,25,45,65,85]
category_4= [15,35,55,75,95]

data = [category_1, category_2, category_3, category_4]

# Create a boxplot
plt.boxplot(data,patch_artist=True,
            boxprops=dict(facecolor='pink',color='black'),
            medianprops=dict(color='red'))

# Add title and labels
plt.title('Boxplot')
plt.xlabel('Categories')
plt.ylabel('Values')

# Show the plot
plt.show()
```
