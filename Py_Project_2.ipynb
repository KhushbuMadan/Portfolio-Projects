```python
from google.colab import files
import pandas as pd
import matplotlib.pyplot as plt

# To Upload the file
uploaded = files.upload()

# To Load the dataset
df = pd.read_csv('vgsales.csv')
print(df.head())
```

```python
import matplotlib.pyplot as plt
import pandas as pd

#line chart
plt.plot(df['Year'],df['EU_Sales'])

#To show line chart
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# To create bins for years
bins = [1980, 1990, 2000, 2010, 2020]
labels = ['1980s', '1990s', '2000s', '2010s']

# To create a new column for year bins
df['Year_Bin'] = pd.cut(df['Year'], bins=bins, labels=labels, right=False)

# To group data by year bins and aggregate EU_Sales
grouped_data = df.groupby('Year_Bin')['EU_Sales'].sum().reset_index()

# Plot the aggregated data
plt.plot(grouped_data['Year_Bin'], grouped_data['EU_Sales'])

# Add title and labels
plt.title('EU Sales by Decade')
plt.xlabel('Decade')
plt.ylabel('EU Sales (in millions)')

# Show line chart
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# To create bins for years
bins = [1980, 1990, 2000, 2010, 2020]
labels = ['1980s', '1990s', '2000s', '2010s']

# To create a new column for year bins
df['Year_Bin'] = pd.cut(df['Year'], bins=bins, labels=labels, right=False)

# To group data by year bins and aggregate Global_Sales
gd = df.groupby('Year_Bin')['Global_Sales'].sum().reset_index()

# Plot the aggregated data in a Scatter Plot
plt.scatter(gd['Year_Bin'], gd['Global_Sales'])

# Add title and labels
plt.title('Global Sales by Decade')
plt.xlabel('Decade')
plt.ylabel('Global Sales (in millions)')

# adding labels to each data point
for i, v in enumerate(gd['Global_Sales']):
   plt.text(i, v+60, str(v),ha='center')

# Show Scatter chart
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# To group data by Genre and aggregate Global_Sales
gd = df.groupby('Genre')['Global_Sales'].sum().reset_index()

# Plot the aggregated data in a Bar
plt.bar(gd['Genre'], gd['Global_Sales'], color='green')

# Add title and labels
plt.title('Global Sales by Genre')
plt.xlabel('Genre')
plt.ylabel('Global Sales (in millions)')

#To rotate the labels on x axis
plt.xticks(rotation=90)

# adding labels to each bar
for i, v in enumerate(gd['Global_Sales']):
   plt.text(i, v+10, str(v),ha='center')

# Show Bar chart
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# To group data by Genre and aggregate Global_Sales
gd = df.groupby('Genre')['Global_Sales'].sum().reset_index()

# Plot the aggregated data in a Pie
plt.pie(gd['Global_Sales'], labels= gd['Genre'], autopct='%1.0f%%')

# Add title and labels
plt.title('Global Sales by Genre')

#To add a legend
plt.legend(loc=(0.9,0.9))

# Show Pie chart
plt.show()
```
